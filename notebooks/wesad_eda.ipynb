```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path

from src.config import load_config
from src.data_loading import load_dataset
from src.preprocessing import preprocess_dataset
```

```
cfg = load_config("configs/wesad_stress.yaml")
cfg
```

```
raw = load_dataset(
    cfg["dataset"]["name"],
    cfg["dataset"]["raw_path"]
)
print("Subjects loaded:", raw.keys())
```

```
processed = preprocess_dataset(
    raw,
    cfg["preprocessing"],
    cfg["dataset"]["interim_path"]
)

sid = list(processed.keys())[0]
sub = processed[sid]
sub.keys()
```

```
plt.figure(figsize=(15, 4))
plt.plot(sub["ecg"][:2000])
plt.title(f"{sid} ECG (first 2000 samples)")
plt.show()

plt.figure(figsize=(15, 4))
plt.plot(sub["eda"][:2000])
plt.title(f"{sid} EDA (first 2000 samples)")
plt.show()
```

```
import collections

labels_all = []
for sid in processed:
    labels_all.extend(processed[sid]["label"])

cnt = collections.Counter(labels_all)
cnt
```

```
plt.figure(figsize=(6, 4))
sns.barplot(x=list(cnt.keys()), y=list(cnt.values()))
plt.title("Label Distribution")
plt.xlabel("Label")
plt.ylabel("Count")
plt.show()
```
